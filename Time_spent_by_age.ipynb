I found a interesting dataset of how much time a person in the USA spends with friends, family, sopuse, children and alone. I found this data on ourworldindata.org.

The average time spent with others is measured in minutes per day, and shown by the age of the
respondent. This is based on averages from surveys between 2009 and 2019.

```python
#importing the necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('time-spent-with-relationships-by-age-us.csv')
data.head(10)
```

```python
data.tail(10)
```

```python
data.info()
```

```python
data.drop(['Entity','Code'],axis=1,inplace=True)
```

```python
data.head()
```

```python
# Renaming the column names for convinience
data.rename(columns={'Time spent alone, by age of respondent (United States)':'Alone', 'Time spent with friends, by age of respondent (United States)':'Friends', 'Time spent with children, by age of respondent (United States)':'Children','Time spent with with parents, siblings and other family, by age of respondent (United States)':'Family Members','Time spent with partner, by age of respondent (United States)':'Partner','Time spent with coworkers, by age of respondent (United States)':'Coworkers'},inplace=True)
data.head()
```

```python
data.isnull().sum()
```

```python
#setting the figure size
plt.rcParams["figure.figsize"] = (16,8)
plt.grid(axis='y', linestyle='--', alpha=0.8)

#plotting the data
plt.plot(data['Year'],data['Alone'],label='Alone',marker='o')
plt.plot(data['Year'],data['Friends'],label='Friends',marker='o')
plt.plot(data['Year'],data['Children'],label='Children',marker='o')
plt.plot(data['Year'],data['Family Members'],label='Family Members',marker='o')
plt.plot(data['Year'],data['Partner'],label='Partner',marker='o')
plt.plot(data['Year'],data['Coworkers'],label='Coworkers',marker='o')

# adding the labels
plt.xlabel('Age in Years',fontsize=20)
plt.ylabel('Time spent per day in minutes',fontsize=20)
plt.suptitle('Who Americans spend their time with, by age',fontsize=20)
plt.title('The average time spent with others is measured in minutes per day, and shown by the age of the respondent.\n This is based on averages from surveys between 2009 and 2019.',fontsize=12)
plt.ylim(0, 600)
plt.xlim(10, 90)
plt.legend(loc='upper left', handlelength=4, borderpad=2,labelspacing=1.5,fontsize=12)

#displaying the plot
plt.show()
```
